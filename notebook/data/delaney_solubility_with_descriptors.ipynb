**Delaney Solubility with Descriptors**

# Load Dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('/content/delaney_solubility_with_descriptors.csv')
df
```

# Data Preparation

### X and *y*

```python
y = df['logS']
y
```

```python
X = df.drop('logS', axis=1)
X
```

# Data Splitting train and test

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=100)
```

```python
X_train
```

```python
X_test
```

# Model Building



### 1. Linear Regression

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, y_train) # 80% of data training
```

#### Applying the model to make a prediction

```python
y_lr_train_pred = lr.predict(X_train)
y_lr_test_pred = lr.predict(X_test)

print(y_lr_train_pred)
```

```python
y_lr_test_pred
```

#### Evaluate the model performance

```python
from sklearn.metrics import mean_squared_error, r2_score

lr_train_mse = mean_squared_error(y_train, y_lr_train_pred)
lr_train_r2 = r2_score(y_train, y_lr_train_pred)

lr_test_mse = mean_squared_error(y_test, y_lr_test_pred)
lr_test_r2 = r2_score(y_test, y_lr_test_pred)


print('LR MSE (Train): ', lr_train_mse)
print('LR R2 (Train): ', lr_train_r2)
print('LR MSE (Test): ', lr_test_mse)
print('LR R2 (Test): ', lr_test_r2)
```

```python
lr_result = pd.DataFrame(['Linear regresion', lr_train_mse, lr_train_r2, lr_test_mse, lr_test_r2]).transpose()
lr_result.columns = ['Method', 'Training MSE', 'Training R2', 'Test MSE', 'Test R2']

lr_result
```

### 2. Random Forest

#### Training the model

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(max_depth=2, random_state=100)
rf.fit(X_train, y_train)
```

#### Applying the model to make a prediction

```python
y_rf_train_pred = rf.predict(X_train)
y_rf_test_pred = rf.predict(X_test)
```

Evaluate the model preformance

```python
from sklearn.metrics import mean_squared_error, r2_score

rf_train_mse = mean_squared_error(y_train, y_rf_train_pred)
rf_train_r2 = r2_score(y_train, y_rf_train_pred)

rf_test_mse = mean_squared_error(y_test, y_rf_test_pred)
rf_test_r2 = r2_score(y_test, y_rf_test_pred)

print('RF MSE (Train): ', rf_train_mse)
print('RF R2 (Train): ', rf_train_r2)
print('RF MSE (Test): ', rf_test_mse)
print('RF R2 (Test): ', rf_test_r2)
```

```python
rf_result = pd.DataFrame(['Random forest', rf_train_mse, rf_train_r2, rf_test_mse, rf_test_r2]).transpose()
rf_result.columns = ['Method', 'Training MSE', 'Training R2', 'Test MSE', 'Test R2']
rf_result
```

# Model Comparison


```python
df_models = pd.concat([lr_result, rf_result], axis=0).reset_index(drop=True)
df_models
```

# Data Visualization of prediction results

```python
plt.figure(figsize=(5, 5))
sns.scatterplot(x=y_train, y=y_lr_train_pred, c='#7cb342', alpha=0.3)

z = np.polyfit(y_train, y_lr_train_pred, 1)
p = np.poly1d(z)

plt.plot(y_train, p(y_train), '#ff7f00')
plt.ylabel('Predict Logs')
plt.xlabel('Experimental Logs')
plt.show()
```
